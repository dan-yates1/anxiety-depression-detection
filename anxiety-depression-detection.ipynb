```python
# read data into pandas dataframe

df = pd.read_csv('../input/well-being-dataset/well-being.csv')
df.head()
```

# Prepare dataset

```python
# check for null values

df.isnull().sum()
df
```

```python
# remove null values

df = df.dropna()
df
```

```python
# replace 'yes' and 'no' to 1 and 0

df.replace(('yes', 'no'), (1, 0), inplace=True)
```

```python
# replace objects with numerical values

df.replace(('heavily', 'frequently', 'regularly', 'occasionally', 'no'), (4, 3, 2, 1, 0), inplace=True)
```

```python
# take out columns which are not relevant for this project

df2 = df.iloc[:, [9,17,25,28,55,57,60,61]]
df2
```

# Build models

```python
# split target variables into seperate array

X = df2
anxiety_y = df['Anxiety symptoms']
depression_y = df['Depressive symptoms']

X2 = df.loc[:, df.columns != 'Anxiety symptoms']
```

```python
depression_y.value_counts()
```

```python
# evaluate performance function

from sklearn import metrics

def evaluate_accuracy(model, X_test, y_test):
    '''return accuracy score of the model'''
    y_pred = model.predict(X_test)
    return(model.score(X_test, y_test))
```

```python
# build logistic regression model function

from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

def build_lr_model(X, y):
    '''return the model and accuracy score'''
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
    model = LogisticRegression()
    model.fit(X_train, y_train)
    return model, evaluate_accuracy(model, X_test, y_test)

anxiety_model = build_lr_model(X, anxiety_y)
depression_model = build_lr_model(X, depression_y)

print("Accuracy score for anxiety logistic regression model: {}".format(anxiety_model[1]))
```

```python
# build mlp classifier

from sklearn.neural_network import MLPClassifier
from sklearn.neural_network import MLPRegressor

X_train, X_test, y_train, y_test = train_test_split(X, anxiety_y, test_size=0.3, random_state=0)

mlp = MLPClassifier(hidden_layer_sizes=(8,8,8), activation='relu', solver='adam', max_iter=500)
mlp.fit(X_train,y_train)
```

```python
# evaluate model

from sklearn.metrics import classification_report,confusion_matrix\

predict_train = mlp.predict(X_train)
predict_test = mlp.predict(X_test)

print(confusion_matrix(y_train,predict_train))
print(classification_report(y_train,predict_train))

print(confusion_matrix(y_test,predict_test))
print(classification_report(y_test,predict_test))
```

```python
# build and compile CNN model

import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.utils.np_utils import to_categorical

y_train = to_categorical(y_train)
y_test = to_categorical(y_test)

model = Sequential()
model.add(Dense(500, activation='relu', input_dim=8))
model.add(Dense(100, activation='relu'))
model.add(Dense(50, activation='relu'))
model.add(Dense(2, activation='softmax'))

model.compile(optimizer='adam', 
              loss='categorical_crossentropy', 
              metrics=['accuracy'])

model.fit(X_train, y_train, epochs=10)
```

```python
# evaluate model

pred_train= model.predict(X_train)
scores = model.evaluate(X_train, y_train, verbose=0)
print('Accuracy on training data: {}% \n Error on training data: {}'.format(scores[1], 1 - scores[1]))   
 
pred_test= model.predict(X_test)
scores2 = model.evaluate(X_test, y_test, verbose=0)
print('Accuracy on test data: {}% \n Error on test data: {}'.format(scores2[1], 1 - scores2[1]))    
```
