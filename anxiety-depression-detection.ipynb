```python
# read data into pandas dataframe

df = pd.read_csv('../input/well-being-dataset/well-being.csv')
df.head()
```

# Clean dataset

```python
# take out columns which are not relevant for this project

df2 = df.iloc[:, [9,17,25,28,52,53,54,55,57,60,61]]
df2
```

```python
# check for null values

df2.isnull().sum()
df2
```

```python
# remove null values

df2 = df2.dropna()
df2
```

```python
# replace 'yes' and 'no' to 1 and 0

df2.replace(('yes', 'no'), (1, 0), inplace=True)
df2
```

```python
# replace objects with numerical values

df2.replace(('heavily', 'frequently', 'regularly', 'occasionally', 'no'), (4, 3, 2, 1, 0), inplace=True)
df2
```

```python
# split target variable into seperate array

anxiety_X = df2.loc[:, df2.columns != 'Anxiety symptoms']
anxiety_y = df2['Anxiety symptoms'].to_numpy()
```

```python
anxiety_X['Cigarette smoker (5 levels)'].unique()
```

```python
# build logistic regression model

from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(anxiety_X, anxiety_y, test_size=0.3, random_state=0)
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
# evaluate performance

from sklearn import metrics

def evaluate_accuracy(model, X_test, y_test):
    y_pred = model.predict(X_test)
    return(model.score(X_test, y_test))

print('Accuracy of model: {:.2f}'.format(evaluate_accuracy(model, X_test, y_test)))
```
